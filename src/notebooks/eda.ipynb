```python
# lets scrape all the matches from previous season for all teams
# number of matches should be 20*19*2 = 760 matches
# we make a match ID data frame
# match ID: team1_team2_date_homeaway(for team1) where team1 is lexicographically smaller than team2
# for each match, we will create a dataframe of players who played and their statistics paths
match_list = {}
for team in teams_24_25[0]:
    team_path = f"{datafiles[0]}/{team}"
    players = glob(f"{team_path}/*")
    print(team_path, len(players))
    for player in players:
        player_name = os.path.basename(player)
        json_files = glob(f"{player}/summary.j*")
        # use on summary json files to get match info
        for jf in json_files:
            jf_name = os.path.basename(jf)
            if jf_name.endswith('.json') and (not jf_name.endswith('all.json')):
                with open(jf, 'r') as f:
                    match_data = json.load(f)
                for match in match_data:
                    # we need to create match ID
                    date = match['Date'].replace('-', '')
                    opponent = match['Opponent']
                    squad = match['Squad']
                    round = match['Round'].split(' ')[1]
                    venue = match['Venue']  # Home or Away for squad
                    result = match['Result'].split(' ')[0]
                    scoreline = match['Result'].split(' ')[1]  # W/D/L for squad scoreline
                    # order squad and opponent lexicographically
                    team1, team2 = sorted([squad, opponent])
                    # if team 1 is squad and venue is Home, then homeaway is H
                    if team1 == squad:
                        homeaway = 'H' if venue == 'Home' else 'A'
                    else:
                        homeaway = 'A' if venue == 'Home' else 'H'
                        # reverse scoreline
                        #print(scoreline, scoreline.split('–'))
                        scoreline = f"{scoreline.split('–')[1]}-{scoreline.split('–')[0]}"
                        result = 'W' if result == 'L' else 'L' if result == 'W' else 'D'

                    match_id = f"{team1}_{team2}_{round}_{date}_{homeaway}"
                    if match_id not in match_list:
                        match_list[match_id] = {'team1': team1, 'team2': team2, 'date': date, 'round': round,
                                                'homeaway': homeaway,
                                                'team1_players': [], 'team2_players': [],
                                                'result': result, 'scoreline': scoreline}
                    else:
                        # verify consistency
                        assert match_list[match_id]['team1'] == team1
                        assert match_list[match_id]['team2'] == team2
                        assert match_list[match_id]['date'] == date
                        assert match_list[match_id]['round'] == round
                        assert match_list[match_id]['homeaway'] == homeaway
                    # add player statistics path to appropriate team
                    # check if player played in the match
                    time_played = match['Min']
                    if time_played.lower() == "match report":
                        continue
                    if squad == team1:
                        match_list[match_id]['team1_players'].append({'player_name': player_name, 'stats_path': jf})
                    else:
                        match_list[match_id]['team2_players'].append({'player_name': player_name, 'stats_path': jf})
                    # for each match, we maintain a list of player statistics paths for each team
```

```python
# DNP
all_teams = list(set(teams_24_25[0]).difference(["Leeds United", "Sunderland", "Burnley"]))
```

```python
did = set()
# make dummy match ID
for i in all_teams:
    for j in all_teams:
        if i==j:
            continue
        if i < j:
            match_id = f"{i}_{j}"
            print(match_id)
        else:
            match_id = f"{j}_{i}"
            print(match_id)
        did.add(match_id)
```

```python
# count matches for each id in did
match_id_counts = {}
for mid in did:
    count = 0
    for match_id in match_list.keys():
        if match_id.startswith(mid):
            count += 1
    match_id_counts[mid] = count
```

```python
# print !=2
{mid: count for mid, count in match_id_counts.items() if count != 2}
```

```python
# sum all counts
sum(match_id_counts.values())
```

```python
# print all match IDs not in did
for match_id in match_list.keys():
    if not any(match_id.startswith(mid) for mid in did):
        print(match_id)
```

```python
# for all match IDs, check length of squad players
Ids_to_exclude = []
for match_id, match_info in match_list.items():
    team1_players = match_info['team1_players']
    team2_players = match_info['team2_players']
    if len(team1_players) < 11 or len(team2_players) < 11:
        print(f"Match ID: {match_id} has less than 11 players for a team.")
        Ids_to_exclude.append(match_id)
        print(f"Team1: {match_info['team1']} Players: {len(team1_players)}")
        print(f"Team2: {match_info['team2']} Players: {len(team2_players)}")
```

```python
len(match_list)-len(Ids_to_exclude)
# train set includes all matches except those in Ids_to_exclude
train_set = {mid: match_info for mid, match_info in match_list.items() if mid not in Ids_to_exclude}
```

```python
# check if team names in train set are only from all_teams
train_teams = set()
for match_id, match_info in train_set.items():
    train_teams.add(match_info['team1'])
    train_teams.add(match_info['team2'])
```
